# 4. 프로그래밍 구조

## 1. 프로그래밍

* 컴퓨터를 이용해 문제를 해결하기 위해 주어진 절차를 문법에 맞게 작성하는 과정
* 데이터 분석을 진행하다 보면 R에서 제공하는 함수만으로는 해결이 되지 않는 경우가 있으며, 이에 프로그래밍 기법을 활용하면 보다 효율적으로 문제를 해결할 수 있음.

## 2. 조건문

* 조건에 따라 특정한 동작을 실행하도록 하는 프로그래밍 명령어

### A. if문

* 조건이 참(`TRUE`)이면 코드블록을 실행하고, 거짓(`FALSE`)이면 코드블록을 실행하지 않고 다음 명령문으로 넘어가는 조건문
* 코드블록 : 중괄호 `{}`로 구분된 코드로, 여러 명령문을 하나로 묶은 실행 단위

```
if (Conditional Statement) {
    조건이 참일 때 실행할 명령문
}
```

> 약간 C 조건문이랑 스타일이 비슷함.

### B. if-else 문

* 조건이 참이면 첫 번째 코드블록을 실행, 거짓이면 두 번째 코드블록을 실행하는 조건문
* if-else문에서 else는 반드시 if문의 코드블록을 닫는 `}`와 같은 줄에 작성해야함;;

### C. `ifelse` 함수

* 조건에 따라 두 값 중 하나를 선택하는 경우, ifelse() 함수를 이용하면 편리함
> SQL, 엑셀 느낌

### D. else if 문

* 첫 번째 조건이 거짓(FALSE)이면 두 번째 조건을 검사하고, 두 번째 조건이 참(TRUE)이면 해당 코드블록을 실행하는 조건문
* 여러 조건을 순차적으로 검사할 때 사용함
* 약자로 `elif`를 쓰지 않고, `else if`를 그대로 써줘야 함. `else if`라는 게 따로 있는 게 아니라 그냥 `else`로 분기해 왔을 때, `if`문을 한 번 더 써주는 그런 방식으로 동작하는 듯
