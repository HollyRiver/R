# 3. 데이터와 자료구조

## 1. 데이터 이해

### A. 형태에 따른 분류

|구분|내용|
|:-:|:-|
|정형 데이터|- 일정한 규칙을 가지고 체계적으로 정리된 데이터, 그 자체로 의미 해석이 가능. <br> - 내부 시스템에 의한 데이터를 수집하기 쉽고, 형식을 가지고 있어 처리하기 쉬움|
|반정형 데이터|- XML, HTML 등의 형태로 존재하는 데이터, 메타데이터나 구조정보 등을 포함 <br> - API 형태로 제공되므로 데이터 처리 기술이 필요|
|비정형 데이터|- 텍스트 문서, 이미지, 동영상 등과 같이 구조화되지 않은 형태의 데이터|

### B. 데이터 구조화

* 데이터를 효과적으로 분석하려면 일정한 형태로 정리되어 있어야 하며, 대표적인 형태가 바로 테이블(table)임
* 데이터를 논리적으로 정리하여 표 형태로 만드는 것을 **데이터 구조화** 또는 **정형화**라고 함
    * 구조화된 데이터는 쉽게 입력되고, 저장·분류할 수 있음
    * 비정형 데이터는 분석을 위해서는 정형 데이터로 변환해야 함

### C. 테이블(table)

* 데이터의 저장형태로, 동질적인 데이터 값인 열과, 하나의 데이터 집합인 행으로 구성됨

> 각각의 열에는 데이터의 값이 동질적임 : 변수
>
> 하나의 행은 한 객체의 데이터 집합이 들어있음 : 관측값, 레코드
>
> 거의 모든 테이블은 각각의 열에 대해서 변수의 이름이 뭔지 헤더 Header를 포함하고 있음.

## 2. 자료구조

* 문제해결을 위해 프로그램이 효율적으로 접근, 변경, 연산이 가능하도록 만든 값들의 구조화된 집합
> 여러 개의 값들을 저장할 때, 어떻게 저장할 수 있는지 그 구조에 대한 것

* R은 벡터, 행렬, 배열, 리스트, 데이터프레임의 5개 기본 자료구조를 제공함

|구분|1차원|2차원|n차원|
|:-:|:-:|:-:|:-:|
|동질적|벡터|행렬|배열|
|이질적|리스트|데이터프레임|-|

> 자료구조에 들어가는 데이터의 자료형이 동질적/이질적인지를 규명한다.

> 데이터프레임은 단일 열에 한정하여 자료형이 동질하다. 데이터프레임은 일반적으로 데이터를 저장하는 형태, 테이블의 형태이다.
>
> 리스트의 경우에는 각 원소들의 자료형이나 자료구조 자체가 달라도 무방하다.
