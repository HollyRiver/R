## 1. 데이터의 이해

### A. 데이터의 시대

* 정보화 시대 -> 데이터의 시대
* 우리를 둘러싼 모든 것들이 데이터 소스와 연결되고, 생활의 많은 부분이 데이터에 의존하여 영위함
* 4차 산업혁명
> 인공지능, 빅데이터라는 핵심기술 -> 결국엔 데이터.
> 향후 데이터를 잘 다룰 줄 아는 기업과 개인이 경쟁에서 우위를 점할 것임.

### B. 데이터 과학

* 데이터를 통해 실제 현상을 이해하고 분석하기 위해 통계학, 데이터 분석, 기계학습과 연관된 방법론을 통합하는 개념
* 컴퓨터 과학 역량, 수학과 통계학 지식, 분야별 전문지식의 교집합이라 소개됨
* 사용하고자 하는 분야에 대한 이해도 중요하다. Domain Knowleage

&nbsp;

* Hadley Wickham이 제시한 데이터 과학의 과정 : 이 양반이 R을 개발함



1. 데이터 가져오기 : 파일/데이터베이스 등을 분석 시스템으로 가져옴
2. 정리하기(tidy) : 가져온 데이터를 일관된 형태로 재구조화. 이 단계를 통해 정리한 데이터를 tidy data라 명명함
3. 변환하기(transform) : tidy data에서 필요로 하는 새로운 값을 구하거나, 분석에 필요한 데이터로 변환
4. 시각화(visualization) : 데이터에서 의미를 도출하는 과정, 예상하지 못한 것을 보여주거나 한 차원 높은 질문을 이끌어낼 수 있음
5. 모델링 : 데이터를 통해 수립한 모델을 이용하여 예측 등의 과정을 수행. 위 셋은 반복될 수 있음.
6. 소통하기 : 데이터에서 찾아낸 세상의 모습을 다른 사람들에게 공유

우리는 위 과정을 모두 다 할겁니다. 네.

## 2. R 및 R Studio 설치

### A. R

* 통계학자가 만듦. AT&T에서 개발한 S를 향상시켜 뉴질랜드 오클랜드 대학에서 R이 개발됨
* 무료로 제공되는 오픈소스 소프트웨어임
* 패키지를 통한 확장성이 뛰어나, 다른 R 사용자들이 개발한 새로운 분석기법을 자유롭게 추가할 수 있음. 다른 소프트웨어에 비해 최신 이론이나 기법을 사용하기 쉬움
* 다양한 통계분석 및 뛰어난 그래픽 기능
* 편리한 도움말 기능
* 프로그램 명령어를 직접 입력하는 방식

### D. 화면 구성

* 기본적으로 네 개의 탭으로 나뉜다.
* R은 인터프리터 언어로, 콘솔 창에서 한줄한줄 입력하여 연산이 가능하다.
* 일반적으로 스크립트 란에 콘솔 창에 실행하고 싶은 명령어를 입력한 뒤, Crtl+Enter / Alt+Enter로 실행이 가능하다.
* 한 번에 실행하고 싶을 경우, 실행하고 싶은 영역을 드래그하고 실행하면 된다. 또는 Crtl+Alt+R로 전체 실행 가능

---

* Environment 창에선 현재 사용하고 있는 변수를 표시한다.
* History 창에선 콘솔에서 실행되었던 내역을 표시한다.
* 시험볼 때, History가 남아있을 수 있으므로, 히스토리 창을 전부 비워놓고 시행한다.

* R studio에서 스크립트를 실수로 꺼도 사라지지 않는다.
* 시험볼 때, 기존에 남아있던 스크립트도 전부 꺼줘야 함.
