## 3. 데이터 테이블 전처리

**전체 코드**

```python
### 총 정리
## 교육 중분류명이 없는 튜플 제거
## 강의코드 인덱스를 강의URL주소에서 따로오고 정수형으로 변경, 추출된 열은 제거
featured_kocw = kocw[["교육 대분류명", "교육 중분류명", "공개강의명", "제공대학명", "교수명", "강의학기", "강의URL주소"]].loc[kocw["교육 중분류명"].isna().map(lambda x : False if x else True), :]\
.assign(강의코드 = lambda _df : _df.강의URL주소.str.split("=").str[-1].astype("int"))\
.set_index("강의코드").reset_index().drop("강의URL주소", axis = 1).reset_index(drop = True)

## 테이블 분리
professor = featured_kocw[["교수명", "교육 대분류명", "제공대학명", "강의코드"]].rename({"교수명":"담당교수명", "교육 대분류명":"담당분야"}, axis = 1)  ## 담당교수
university = featured_kocw[["제공대학명"]]  ## 제공대학
course = featured_kocw[["강의코드", "공개강의명", "교육 중분류명", "강의학기", "제공대학명"]].rename({"공개강의명":"강의명", "교육 중분류명":"교육분류"}, axis = 1)  ## 강의
charge = featured_kocw[["강의코드", "교수명"]].rename({"교수명":"담당교수명"}, axis = 1)  ## 담당

## 강의학기 -> 강의년도 + 강의학기
featured_course = course.assign(강의년도 = lambda _df : _df.강의학기.str.split(" ").str[0].str[:-1].astype(int), 강의학기 = lambda _df : _df.강의학기.str.split(" ").str[1].str[0].astype(int))\
.set_index(["강의코드", "강의명", "교육분류", "강의년도", "강의학기"]).reset_index()

## 정규화 과정 중 문제가 발생하여 일부 데이터 임의로 수정
ind = featured_course[["강의명", "교육분류"]].drop_duplicates().강의명.drop_duplicates().index
temp = featured_course[["강의명", "교육분류"]].loc[ind].reset_index(drop = True)
temp_dict = {i:j for i, j in zip(temp.강의명, temp.교육분류)}


##-----------------------교수명 속성 처리 및 교수코드 생성----------------------##

## professor table
df = professor.drop("강의코드", axis = 1).drop_duplicates()\
.assign(temp = lambda _df : _df.담당교수명.map(lambda x : x.split("+ ") if "+ " in x else x.split("+")))\
.reset_index(drop = True)\
.set_index(["담당교수명", "담당분야", "제공대학명"]).apply(lambda x : x.explode()).reset_index().drop("담당교수명", axis = 1)\
.rename({'temp' : "담당교수명"}, axis = 1)\
#빈 칸 없애기
df_empty = df[df['담당교수명'] == ""].index
df.drop(df_empty, inplace=True)

df2 = df[df['담당교수명'].str.contains('외.*인|외.*명| 외', regex=True)]  ## 문제 있는 열만 추출(외 몇명)
temp = df2['담당교수명'].str.split(' 외').str[0].str.split('외').str[0]

featured_df = df

for i in list(temp.index) :
    featured_df.loc[i,"담당교수명"] = temp.loc[i]

featured_df["담당교수명"] = featured_df["담당교수명"].str.split('(').str[0].str.strip()
featured_df = featured_df.drop_duplicates().reset_index().rename({"index":"교수코드"}, axis = 1)  ## professor table(최종)

## change table(담당 테이블)
new_pro= professor.drop_duplicates()\
.assign(temp = lambda _df : _df.담당교수명.map(lambda x : x.split("+ ") if "+ " in x else x.split("+")))\
.reset_index(drop = True)\
.set_index(["담당교수명", "담당분야", "제공대학명", "강의코드"])\
.apply(lambda x : x.explode()).reset_index().drop("담당교수명", axis = 1).rename({'temp' : "담당교수명"}, axis = 1)

new_pro_empty = new_pro[new_pro['담당교수명'] == ""].index
new_pro.drop(new_pro_empty, inplace=True)

new_pro2 = new_pro[new_pro['담당교수명'].str.contains('외.*인|외.*명| 외', regex=True)]
new_temp = new_pro2['담당교수명'].str.split(' 외').str[0].str.split('외').str[0]

featured_df2 = new_pro.copy()

for i in list(new_temp.index) :
    featured_df2.loc[i,"담당교수명"] = new_temp.loc[i]


featured_df2 = new_pro.copy()
for i in list(new_temp.index):
    featured_df2.loc[i, "담당교수명"] = new_temp.loc[i]
featured_df2['담당교수명'] = featured_df2['담당교수명'].str.split('(').str[0].str.strip()

featured_df2 = featured_df2.drop_duplicates()

## mapping
fet2 = featured_df2.assign(구분 = lambda _df : _df.담당교수명 + '-' + _df.담당분야 + '-' + _df.제공대학명)
fet2 ## 17000행 테이블 코드

featured_map = featured_df.assign(구분 = lambda _df : _df.담당교수명 + '-' + _df.담당분야 + '-' + _df.제공대학명)
mapping_dict = featured_map.set_index('구분')['교수코드'].to_dict()

fet2['mapping'] = fet2['구분'].map(mapping_dict)
charge = fet2.rename({"mapping":"교수코드"}, axis = 1)[["교수코드", "강의코드"]].reset_index(drop = True)  ## charge table(최종)
```
